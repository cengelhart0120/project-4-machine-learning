```python
# Calculate the average of "G1" and "G2" features
average_MIDGRADE = (df['G1'] + df['G2']) / 2

# Round the average values to the nearest integer using specific rounding rule
rounded_MIDGRADE = np.where(average_MIDGRADE - np.floor(average_MIDGRADE) >= 0.5, np.ceil(average_MIDGRADE), np.floor(average_MIDGRADE))

# Create a new feature "parental level of education" with rounded integer values
df['MIDGRADE'] = rounded_MIDGRADE.astype(int)

# Print the updated dataset
df
```

```python
df["G3"] = df["G3"].astype(int)
df["passed"] = np.where(df["G3"] >= 10, 1, 0)
```

```python
# mapping strings to numeric values:
def numerical_data():
    df['school'] = df['school'].map({'GP': 0, 'MS': 1})
    df['sex'] = df['sex'].map({'M': 0, 'F': 1})
    df['address'] = df['address'].map({'U': 0, 'R': 1})
    df['famsize'] = df['famsize'].map({'LE3': 0, 'GT3': 1})
    df['Pstatus'] = df['Pstatus'].map({'T': 0, 'A': 1})
    df['Mjob'] = df['Mjob'].map({'teacher': 0, 'health': 1, 'services': 2, 'at_home': 3, 'other': 4})
    df['Fjob'] = df['Fjob'].map({'teacher': 0, 'health': 1, 'services': 2, 'at_home': 3, 'other': 4})
    df['reason'] = df['reason'].map({'home': 0, 'reputation': 1, 'course': 2, 'other': 3})
    df['guardian'] = df['guardian'].map({'mother': 0, 'father': 1, 'other': 2})
    df['schoolsup'] = df['schoolsup'].map({'no': 0, 'yes': 1})
    df['famsup'] = df['famsup'].map({'no': 0, 'yes': 1})
    df['paid'] = df['paid'].map({'no': 0, 'yes': 1})
    df['activities'] = df['activities'].map({'no': 0, 'yes': 1})
    df['nursery'] = df['nursery'].map({'no': 0, 'yes': 1})
    df['higher'] = df['higher'].map({'no': 0, 'yes': 1})
    df['internet'] = df['internet'].map({'no': 0, 'yes': 1})
    df['romantic'] = df['romantic'].map({'no': 0, 'yes' : 1})

# reorder dataframe columns :
    col = df['passed']
    del df['passed']
    df['passed'] = col
    
# Scale
def feature_scaling(df):
    for i in df:
        col = df[i]
        # let's choose columns that have large values
        if(np.max(col)>6):
            Max = max(col)
            Min = min(col)
            mean = np.mean(col)
            col  = (col-mean)/(Max)
            df[i] = col
        elif(np.max(col)<6):
            col = (col-np.min(col))
            col /= np.max(col)
            df[i] = col
```

```python
# define a function to map the midgrade values to a new scale from 1 to 5
def map_midgrade(x):
    if x <= 8:
        return 1
    elif x < 12:
        return 2
    elif x < 16:
        return 3
    elif x < 18:
        return 4
    else:
        return 5


# apply the mapping function to the performance column
df['mid_mapped'] = df['MIDGRADE'].apply(map_midgrade)
```

```python
# Split the data into training and testing sets

X = newdf.drop(["MIDGRADE"], axis=1)
      # "Walc","failures","school","famsize","goout","famsup"], axis=1)
#X = newdf[['MIDGRADE','sex','higher','paid','Medu', 'Fedu', 'internet', 'studytime','famrel','Pstatus','activities','Fjob','reason','nursery','freetime']]
y = newdf['passed']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

X = newdf.drop(columns = "passed", axis = 1)
y= newdf["passed"]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
       
```

```python
# Multiple linear regression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_predictions = lr.predict(X_test)
lr_mse = mean_squared_error(y_test, lr_predictions)
lr_rmse = lr_mse ** 0.5
lr_r2 = r2_score(y_test, lr_predictions)

lr_table = pd.DataFrame({
    'Actual': y_test,
    'Predicted': lr_predictions,
    'RMSE': lr_rmse,
    'R2': lr_r2
})

print('Multiple Linear Regression Table:')
print(lr_table.head())
```

```python
# Random forest regression
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
rf_predictions = rf.predict(X_test)
rf_mse = mean_squared_error(y_test, rf_predictions)
rf_rmse = rf_mse ** 0.5
rf_r2 = r2_score(y_test, rf_predictions)


rf_table = pd.DataFrame({
    'Actual': y_test,
    'Predicted': rf_predictions,
    'RMSE': rf_rmse,
    'R2': rf_r2
})

print('\nRandom Forest Regression Table:')
print(rf_table.head())
```

```python
# Neural network
nn = MLPRegressor(hidden_layer_sizes=(100, 100), max_iter=1000, random_state=42)
nn.fit(X_train, y_train)
nn_predictions = nn.predict(X_test)
nn_mse = mean_squared_error(y_test, nn_predictions)
nn_rmse = nn_mse ** 0.5
nn_r2 = r2_score(y_test, nn_predictions)

nn_table = pd.DataFrame({
    'Actual': y_test,
    'Predicted': nn_predictions,
    'RMSE': nn_rmse,
    'R2': nn_r2
})

print('\nNeural Network Regression Table:')
print(nn_table.head())
```

```python
# Comparison of the models
print('\nComparison of Models:')
models = ['Multiple Linear Regression', 'Random Forest Regression', 'Neural Network Regression']
rmses = [lr_rmse, rf_rmse, nn_rmse]
r2_scores = [lr_r2, rf_r2, nn_r2]

compare_table = pd.DataFrame({
    'Model': models,
    'RMSE': rmses,
    'R2 Score': r2_scores
})

print(compare_table)
```

```python



# Define the impact metric
impact_metric = 'passed'

# Get a list of all categories in the feature
categories = df['MIDGRADE'].unique()

# Calculate the impact metric for each category
for category in categories:
    # Subset data for the category
    subset = df[df['MIDGRADE'] == category]
    
    # Calculate the impact metric for the subset
    impact_metric_category = subset[impact_metric].mean()
    
    # Calculate the impact metric for the entire dataset
    overall_metric = df[impact_metric].mean()
    
    # Calculate the impact ratio
    impact_ratio = impact_metric_category / overall_metric
    
    # Print the impact ratio
    print(f"The impact of the '{category}' category is {impact_ratio:.2f} times the impact of the entire dataset.")
```
